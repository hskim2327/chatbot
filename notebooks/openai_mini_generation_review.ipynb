## 5. 노트북에서 직접 필터링해서 보기

아래 셀을 실행하면 DataFrame으로 보고 싶은 케이스만 골라볼 수 있습니다.


```python
import json
from pathlib import Path
import pandas as pd

RESULT_PATH = Path('../outputs/generation/openai_mini_representative_wrong_30/20260527_105215_gpt4o_mini_answer_samples.jsonl')
rows = [json.loads(line) for line in RESULT_PATH.read_text(encoding='utf-8').splitlines() if line.strip()]

def norm_doc(value):
    import re
    value = str(value or '').lower()
    value = re.sub(r'\.(hwp|hwpx|pdf|docx?)$', '', value)
    value = re.sub(r'[^0-9a-z가-힣]+', '', value)
    return value

def gt_hit(row):
    gts = [norm_doc(v) for v in row.get('ground_truth_docs') or []]
    docs = [norm_doc(v) for v in row.get('retrieved_docs') or []]
    matched = [gt for gt in gts if gt and any(gt in doc or doc in gt for doc in docs)]
    return f'{len(matched)}/{len(gts)}'

def one_line(text, n=160):
    text = ' '.join(str(text or '').split())
    return text if len(text) <= n else text[:n] + '...'

df = pd.DataFrame([
    {
        'id': r.get('id'),
        'type': r.get('type'),
        'difficulty': r.get('difficulty'),
        'question_type': r.get('question_type'),
        'gt_doc_hit': gt_hit(r),
        'question': one_line(r.get('question'), 120),
        'answer': one_line(r.get('answer'), 180),
        'gold': one_line(r.get('ground_truth_answer'), 180),
        'warnings': ', '.join((r.get('guardrails') or {}).get('warnings') or []),
        'prompt_tokens': (r.get('openai_usage') or {}).get('prompt_tokens'),
        'completion_tokens': (r.get('openai_usage') or {}).get('completion_tokens'),
    }
    for r in rows
])

pd.set_option('display.max_colwidth', 220)
df
```

```python
# 정답 문서가 retrieved docs에 없는 케이스만 보기
df[df['gt_doc_hit'].str.startswith('0/')]
```

```python
# guardrail warning이 있는 케이스만 보기
df[df['warnings'].astype(str).str.len() > 0]
```

```python
from IPython.display import Markdown, display

def show_case(qid):
    row = next(r for r in rows if r['id'] == qid)
    docs = '\n'.join(f'- {doc}' for doc in row.get('retrieved_docs') or [])
    evidence = '\n'.join(
        f"- rank {e.get('rank')} | {e.get('filename')} | {e.get('sentence')}"
        for e in (row.get('evidence_sentences') or [])[:8]
    )
    text = f'''## {row.get('id')} | type {row.get('type')} | {row.get('question_type')}

### Question
{row.get('question')}

### Generated Answer
{row.get('answer')}

### Gold Answer
{row.get('ground_truth_answer')}

### Ground Truth Docs
{row.get('ground_truth_docs')}

### Retrieved Docs
{docs}

### Evidence Candidates
{evidence}

### Guardrails
{row.get('guardrails')}
'''
    display(Markdown(text))

# 예시: show_case('Q006')
show_case(df.iloc[0]['id'])
```
